```python
post_year = 1996
year_dict = {}
full_out = []
mag_keys = set()
mag_sep = []
for i in range(1, 13):
    mag_sep.append(i)
#     mag_sep.append(i + 0.5)

for i, row in cereal_df.iterrows():
    year = row['YEAR']
    month = row['MONTH']
    day = row['DAY']

    if year >= post_year:
        lat = row['LATITUDE']
        lon = row['LONGITUDE']
        mag = row['INTENSITY']        
        if not math.isnan(float(lat)) and not math.isnan(mag):
            inj = row['INJURIES']
            death = row['DEATHS']
            if math.isnan(inj):
                inj = 0
            if math.isnan(death):
                death = 0
                
            injdea = inj + death
            for i in range(len(mag_sep)-1):
                if mag >= mag_sep[i] and mag < mag_sep[i+1]:
                    mag_type = str(mag_sep[i])
            mag_keys.add(mag_type)
            year_key = str(year) + '-01-02'
            if year_key not in year_dict:
                year_dict[year_key] = {}
            if mag_type not in year_dict[year_key]:
                year_dict[year_key][mag_type] = 0
                
            year_dict[year_key][mag_type] += 1
            l = ','.join([str(year), lon, lat, str(mag), mag_type, str(injdea)])
            full_out.append(l)
```

```python
with open('../year_mag.csv', 'w') as f:
    l = 'date'
    for mag_key in mag_keys:
        l += ',' + mag_key
    f.write(l + '\n')
    
    for year_key in year_dict:
        l = year_key
        for mag_key in mag_keys:
            if mag_key not in year_dict[year_key]:
                l += ',0'
            else:
                l += ',' + str(year_dict[year_key][mag_key])
        f.write(l + '\n')
```

```python
with open('../full_out.csv', 'w') as f:
    l = 'id,date,longitude,latitude,magnitude,mag_key,value\n'
    f.write(l)
    for i, l in enumerate(full_out):
        f.write(str(i) + ',' + l + '\n')
```
