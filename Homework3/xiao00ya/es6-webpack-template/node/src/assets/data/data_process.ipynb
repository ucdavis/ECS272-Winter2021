#  aggregate the crime reports according to its type and the time


```python
categories = Counter(table["Category"])
```

```python
categories.keys()
```

```python
category_map = {
"Personal Crimes" : [],
"Property Crimes" : [],
"Inchoate Crimes" : [],
"Statutory Crimes" : [],
"Financial and Other Crimes" : []
}
```

```python
Counter(table["PdDistrict"]).keys()
Counter(table["Time"]).keys()
```

```python
# initialize
categories_dict = {}
for category in categories.keys():
    categories_dict[category] = 0
    
```

```python
len(categories_dict)
```

```python
dates = Counter(table["Date"]).keys()
crime_categories_history = {}
for item in dates:
    crime_categories_history[item] = categories_dict.copy()
    
```

```python
# parse the data
total_count = len(table["PdId"])
print("The number of lines: ", total_count)
for index in tqdm.tqdm_notebook(range(total_count)):
    crime_type = table["Category"][index]
    date = table["Date"][index]
    time = table["Time"][index]
    crime_categories_history[date][crime_type] += 1
    
```

```python
# sorted(crime_categories_history.keys(), key = lambda k: time.strptime(k, "%m/%d/%Y %I:%M:%S %p"))
```

```python
import time 
time_list = list(crime_categories_history.keys())
# time.strptime(d, "%m/%d/%Y %I:%M:%S %p") for d in crime_categories_history.keys()
# index_order = sorted(range(len(crime_categories_history.keys())), key = lambda k: time.strptime(time_list[k], "%m/%d/%Y %I:%M:%S %p"))
time_list_ordered = sorted(crime_categories_history.keys(), key = lambda k: time.strptime(k, "%m/%d/%Y %I:%M:%S %p"))

crime_history_table = pd.DataFrame()
cates = OrderedDict()
for cat in categories.keys():
    cates[cat] = []
    
for time in time_list_ordered:
    report = crime_categories_history[time]
#     print("date", " ", date, " ", report)
    for key in report.keys():
        cates[key].append(report[key])
```

```python
# convert time format
time_list_update = []
for time in time_list_ordered:
    first_set = time.split()[0]
    month, day, year = first_set.split("/")
    time_list_update.append(year+"-"+month+"-"+day)
```

```python
crime_history_table["date"] = time_list_update
for cat in cates.keys():
    crime_history_table[cat] = cates[cat]
```

```python
Counter(table["DayOfWeek"]).keys()
```

```python
crime_day_distribution = {}
for category in categories.keys():
    crime_day_distribution[category] = {'Monday':0, 'Tuesday':0, 'Wednesday':0,  'Thursday':0, 'Friday':0, 'Saturday':0,'Sunday':0, }
```

```python
total_count = len(table["PdId"])
print("The number of lines: ", total_count)
for index in tqdm.tqdm_notebook(range(total_count)):
    crime_type = table["Category"][index]
    crime_day = table["DayOfWeek"][index]
    crime_day_distribution[crime_type][crime_day] += 1
    
```

```python
crime_day_distribution
```

```python
# post validation --> valid
total = 0
for key in crime_day_distribution:
    total += sum(crime_day_distribution[key].values())
total == total_count
```

```python
crime_day_table = pd.DataFrame()
crime_list = []
day_list = []
value_list = []
for key in crime_day_distribution:
    for day in crime_day_distribution[key]:
        crime_list.append(key)
        day_list.append(day)
        value_list.append(crime_day_distribution[key][day])
```

```python
# check dimension
len(crime_list) == len(day_list) == len(value_list)
```

```python
crime_day_table["category"] = crime_list
crime_day_table["dayOfWeek"] = day_list
crime_day_table["count"] = value_list
```
