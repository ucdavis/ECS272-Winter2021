```python
table = pd.read_csv("/Users/xiao0_0yang/Documents/UCDavisWinter2021/ECS272-Winter2021/Homework2/Templates/es6-webpack-template/node/src/assets/data/sf_crime_zip11.csv")
```

```python
filtered_coarse_categories = []
filtered_categories = []
filtered_zips = []
for i in range(len(table)):
    if math.isnan(table["zip"][i]) == False:
        filtered_categories.append(table["Category"][i])
        filtered_zips.append(int(table["zip"][i]))
        filtered_coarse_categories.append(type_mapping[table["Category"][i]])
#     else:
#         pritn()
```

```python
len(filtered_categories) == len(filtered_coarse_categories) == len(filtered_zips)
```

```python
from collections import Counter
Counter(filtered_coarse_categories)
```

```python
# table
base = {"property crimes":0, "personal crimes":0, "statutory crimes":0, 'weapon laws': 0,
         'warrants': 0,
         'non-criminal': 0,
         'other': 0,
         'financial': 0,
         'secondary codes': 0,
         'suspicious occ': 0,
         'trespass': 0,
         'runaway': 0,
         'family offenses': 0,
         'suicide': 0,
         'liquor laws': 0}
# for coarse_type in crime_type_mapping:
#     base = {}
```

```python
from collections import Counter
# Counter(table["zip"]).keys()
```

```python
zip_count = {}
for i in range(len(filtered_coarse_categories)):
    coarse_type = filtered_coarse_categories[i]
    zip_value = filtered_zips[i]
    if coarse_type in base.keys():
        if zip_value in zip_count:
            zip_count[zip_value][coarse_type] += 1
        else:
            zip_count[zip_value] = base.copy()
            zip_count[zip_value][coarse_type] += 1
```
